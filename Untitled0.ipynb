```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_percentage_error, r2_score

# Load data
df = pd.read_csv('/content/Dataset_Generator_for_DTDC 2.csv')

# Select relevant columns
features = ['Chargeable Wt','Actual Wt','Volumetric Wt',
            'Sender City','Recipient City','Sender Pincode','Receiver Pincode',
            'Mode','Risk Surcharge']
target_tariff = 'Tariff'

model_df = df[features + [target_tariff, 'VAS Charges', 'Total Amount']].copy()

# Drop rows with missing numeric values
model_df.dropna(subset=['Chargeable Wt','Actual Wt','Volumetric Wt',
                        'Sender Pincode','Receiver Pincode',target_tariff], inplace=True)

# Fill missing categorical values with a placeholder. Crucially, this ensures 'Unknown' can be a class.
for col in ['Sender City','Recipient City','Mode','Risk Surcharge']:
    model_df[col] = model_df[col].fillna('Unknown')

# Create a distance proxy
model_df['Pincode_Diff'] = (model_df['Sender Pincode'] - model_df['Receiver Pincode']).abs()

# Encode categorical variables
le_sender = LabelEncoder(); le_recipient = LabelEncoder()
le_mode = LabelEncoder(); le_risk = LabelEncoder()

# Ensure 'Unknown' is always a class for each encoder during fitting
# This handles cases where original data had no NaNs, but 'Unknown' might be used as fallback for user input
for encoder, col_name in [(le_sender, 'Sender City'),
                          (le_recipient, 'Recipient City'),
                          (le_mode, 'Mode'),
                          (le_risk, 'Risk Surcharge')]:
    # Get all unique values from the column, convert to string, add 'Unknown' if not present
    unique_values = model_df[col_name].astype(str).unique().tolist()
    if 'Unknown' not in unique_values:
        unique_values.append('Unknown')
    encoder.fit(unique_values)

# Now transform the actual columns using the fitted encoders
model_df['Sender_City_Enc'] = le_sender.transform(model_df['Sender City'].astype(str))
model_df['Recipient_City_Enc'] = le_recipient.transform(model_df['Recipient City'].astype(str))
model_df['Mode_Enc'] = le_mode.transform(model_df['Mode'].astype(str))
model_df['Risk_Enc'] = le_risk.transform(model_df['Risk Surcharge'].astype(str))

# Prepare features for the tariff model
X = model_df[['Chargeable Wt','Actual Wt','Volumetric Wt',
              'Sender Pincode','Receiver Pincode','Pincode_Diff',
              'Sender_City_Enc','Recipient_City_Enc','Mode_Enc','Risk_Enc']]
y_tariff = model_df[target_tariff]

# Split data into train/test sets
X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X, y_tariff, model_df.index, test_size=0.2, random_state=42)

# Train a Random Forest regressor on tariffs
tariff_model = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
tariff_model.fit(X_train, y_train)

# Predict tariffs for the test set
predicted_tariff = tariff_model.predict(X_test)

# Compute the predicted total amount by adding the actual VAS charge back in
predicted_total = predicted_tariff + model_df.loc[idx_test, 'VAS Charges'].values
actual_total = model_df.loc[idx_test, 'Total Amount'].values

# Calculate accuracy metrics
mape = mean_absolute_percentage_error(actual_total, predicted_total)
accuracy_percentage = 100 * (1 - mape)
r2 = r2_score(actual_total, predicted_total)

print(f"Improved MAPE: {mape:.4f}")
print(f"Improved accuracy (100 × (1 – MAPE)): {accuracy_percentage:.2f}%")
print(f"Improved R² score: {r2:.4f}")
```

```python
def get_current_weather(city, api_key):
    """Return temperature (°C), weather condition and wind speed (kph)."""
    url = f"https://api.weatherapi.com/v1/current.json"
    params = {"key": api_key, "q": city}
    resp = requests.get(url, params=params, timeout=10)
    data = resp.json()
    temp_c = data['current']['temp_c']
    condition = data['current']['condition']['text']
    wind_kph = data['current']['wind_kph']
    return temp_c, condition, wind_kph
```

```python
def get_fuel_price(state, city, rapidapi_key):
    """Fetch latest petrol and diesel prices for a given state and city."""
    url = f"https://daily-fuel-price-india1.p.rapidapi.com/api/fuel/price/{state}/{city}"
    headers = {
        "X-RapidAPI-Key": rapidapi_key,
        "X-RapidAPI-Host": "daily-fuel-price-india1.p.rapidapi.com"
    }
    resp = requests.get(url, headers=headers, timeout=10)
    data = resp.json()
    # The API response body contains a 'payload' object with price fields:contentReference[oaicite:5]{index=5}.
    petrol_price = data['payload']['petrol']
    diesel_price = data['payload']['diesel']
    return petrol_price, diesel_price
```

```python
def get_traffic_data(lat, lon, tomtom_key, zoom=10, style="absolute"):
    """
    Fetch traffic flow information from TomTom.
    Returns current travel time and free-flow travel time in seconds.
    """
    base_url = "https://api.tomtom.com/traffic/services/4/flowSegmentData"
    url = f"{base_url}/{style}/{zoom}/json"
    params = {
        "key": tomtom_key,
        "point": f"{lat},{lon}",
        "unit": "kmph"
    }
    resp = requests.get(url, params=params, timeout=10)
    data = resp.json()

    if 'flowSegmentData' in data:
        flow = data['flowSegmentData']
        current_travel_time = flow['currentTravelTime']  # seconds
        free_flow_travel_time = flow['freeFlowTravelTime']  # seconds
        return current_travel_time, free_flow_travel_time
    else:
        # Handle cases where 'flowSegmentData' is not present (e.g., API error, no data for location)
        print(f"Warning: 'flowSegmentData' not found in TomTom API response for ({lat}, {lon}). Response: {data}")
        # Return default values indicating no impact on traffic (multiplier of 1)
        return 60, 60 # Default to 1 minute current and free-flow time
```

```python
def compute_dynamic_price(base_price,
                          origin_lat, origin_lon,
                          dest_city, dest_state,
                          weather_api_key,
                          fuel_api_key,
                          tomtom_key,
                          baseline_fuel_price=100.0):
    """
    Adjust base price using real-time fuel price, traffic and weather.
    - baseline_fuel_price: reference fuel price used when training the model.
    """
    # Fetch weather at destination city
    temp_c, condition, wind_kph = get_current_weather(dest_city, weather_api_key)
    # Weather multiplier: surcharge when heavy rain/wind or extreme temperature
    if "rain" in condition.lower() or "storm" in condition.lower() or wind_kph > 30:
        weather_multiplier = 1.05
    elif temp_c > 38 or temp_c < 5:
        weather_multiplier = 1.03
    else:
        weather_multiplier = 1.0
 # Fetch latest fuel price (use petrol price as proxy)
    petrol_price, diesel_price = get_fuel_price(dest_state, dest_city, fuel_api_key)
    # Fuel factor is ratio of current petrol price to baseline
    fuel_multiplier = petrol_price / baseline_fuel_price

    # Fetch traffic data at origin coordinates (approximate road congestion)
    current_time, free_flow_time = get_traffic_data(origin_lat, origin_lon, tomtom_key)
    # Traffic factor: 1 when traffic is normal; >1 when travel time is longer than free-flow
    traffic_multiplier = current_time / free_flow_time if free_flow_time else 1.0

    # Combine multipliers to compute dynamic price
    dynamic_price = base_price * fuel_multiplier * traffic_multiplier * weather_multiplier
    return dynamic_price, {
        "weather_condition": condition,
        "temperature_c": temp_c,
        "wind_kph": wind_kph,
        "petrol_price": petrol_price,
        "diesel_price": diesel_price,
        "traffic_ratio": traffic_multiplier
    }
```

```python
import requests

def get_valid_states(rapidapi_key):
    url = "https://daily-fuel-price-india1.p.rapidapi.com/api/fuel/list/states"
    headers = {
        "X-RapidAPI-Key": rapidapi_key,
        "X-RapidAPI-Host": "daily-fuel-price-india1.p.rapidapi.com"
    }
    resp = requests.get(url, headers=headers, timeout=10)
    data = resp.json()
    # If the API returns a 'message' key (e.g. "Too many requests"), raise an error
    if "message" in data and data.get("states") is None:
        raise RuntimeError(f"Fuel API error: {data['message']}")
    return data.get("states", [])

def get_cities_for_state(state, rapidapi_key):
    # Remove spaces from the state name to build the endpoint
    state_param = state.replace(" ", "")
    url = f"https://daily-fuel-price-india1.p.rapidapi.com/api/fuel/list/{state_param}"
    headers = {
        "X-RapidAPI-Key": rapidapi_key,
        "X-RapidAPI-Host": "daily-fuel-price-india1.p.rapidapi.com"
    }
    resp = requests.get(url, headers=headers, timeout=10)
    data = resp.json()
    if "message" in data and data.get("cities") is None:
        raise RuntimeError(f"Fuel API error: {data['message']}")
    return data.get("cities", [])

def get_fuel_price(state, city, rapidapi_key):
    state_param = state.replace(" ", "")
    city_param = city.replace(" ", "")
    url = f"https://daily-fuel-price-india1.p.rapidapi.com/api/fuel/price/{state_param}/{city_param}"
    headers = {
        "X-RapidAPI-Key": rapidapi_key,
        "X-RapidAPI-Host": "daily-fuel-price-india1.p.rapidapi.com"
    }
    resp = requests.get(url, headers=headers, timeout=10)
    data = resp.json()
    # Check for API error messages
    if "payload" not in data:
        raise RuntimeError(f"Fuel API error: {data.get('message', 'Unexpected response')}")
    return data["payload"]["petrol"], data["payload"]["diesel"]
```

```python
import requests

def get_valid_states(rapidapi_key):
    # List all states for the new API
    url = "https://daily-petrol-diesel-lpg-cng-fuel-prices-in-india.p.rapidapi.com/v1/list/india/states"
    headers = {
        "X-RapidAPI-Key": rapidapi_key,
        "X-RapidAPI-Host": "daily-petrol-diesel-lpg-cng-fuel-prices-in-india.p.rapidapi.com"
    }
    resp = requests.get(url, headers=headers, timeout=10)
    data = resp.json()
    # Response contains 'states' as a list of state names or IDs
    return data.get("states", [])

def get_cities_for_state(state, rapidapi_key):
    # State parameter must match the API’s format (lowercase or hyphenated, no spaces)
    state_param = state.lower().replace(" ", "-")
    url = f"https://daily-petrol-diesel-lpg-cng-fuel-prices-in-india.p.rapidapi.com/v1/list/india/{state_param}"
    headers = {
        "X-RapidAPI-Key": rapidapi_key,
        "X-RapidAPI-Host": "daily-petrol-diesel-lpg-cng-fuel-prices-in-india.p.rapidapi.com"
    }
    resp = requests.get(url, headers=headers, timeout=10)
    data = resp.json()
    return data.get("cities", [])

def get_fuel_price(state, city, rapidapi_key):
    """
    Fetch today’s petrol and diesel prices for a given state and city from
    the Daily Petrol, Diesel, LPG & CNG Fuel Prices in India API.
    """
    state_param = state.lower().replace(" ", "-")
    city_param = city.lower().replace(" ", "-")
    url = (f"https://daily-petrol-diesel-lpg-cng-fuel-prices-in-india.p.rapidapi.com"
           f"/v1/fuel-prices/today/india/{state_param}/{city_param}")
    headers = {
        "X-RapidAPI-Key": rapidapi_key,
        "X-RapidAPI-Host": "daily-petrol-diesel-lpg-cng-fuel-prices-in-india.p.rapidapi.com"
    }
    resp = requests.get(url, headers=headers, timeout=10)
    data = resp.json()
    # The API returns 'petrol' and 'diesel' objects nested under a 'fuel' key
    if "fuel" not in data or "petrol" not in data["fuel"] or "diesel" not in data["fuel"]:
        raise RuntimeError(f"Unexpected response: {data}")
    petrol_price = data["fuel"]["petrol"]["retailPrice"]
    diesel_price = data["fuel"]["diesel"]["retailPrice"]
    return petrol_price, diesel_price
```
